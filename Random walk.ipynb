```python
import numpy as np
N=10
```

```python
data=np.random.random((N,))-0.5
```

```python
location=np.cumsum(data)
```

```python
import matplotlib.pyplot as plt
plt.ylabel('Location')
plt.plot(location)
```

```python
plt.plot(data,'ro')
```

```python
n_data=np.zeros(N)
for i in range(0,N):
    i
    if data[i]>0:
        n_data[i]=1
    else:
        n_data[i]=-1
```

```python
n_location=np.cumsum(n_data)
```

```python
import matplotlib.pyplot as plt
plt.ylabel('n_Location')
plt.plot(n_location)
```

```python
xdata=2*(np.random.random((N,))-0.5)
ydata=2*(np.random.random((N,))-0.5)
xlocation=np.cumsum(xdata)
ylocation=np.cumsum(ydata)
plt.ylabel('n_Location')
plt.plot(xdata,ydata)
plt.axis('equal')
plt.xlim((-2,2))
plt.ylim((-2,2))
```
